```python
result = {}
for (year, medal), count in group_medal_count.items():
    if year not in result:
        result[year] = {'G': 0, 'S': 0, 'B': 0}
    if medal == 'Gold':
        result[year]['G'] = count
    elif medal == 'Silver':
        result[year]['S'] = count
    elif medal == 'Bronze':
        result[year]['B'] = count
    #print(year, medal, count)
result
```
